# 데이터에서 불필요한 컬럼 제거

(1) 라이브러리 임포트

```python
import pandas as pd
```

(2) 데이터 로드

```python
df = pd.read_csv(r"C:\Users\user22\Desktop\아파트 값 예측\new_city.csv", encoding='utf-8-sig')
```

(3) 불필요한 컬럼 정의

```python
# 제거할 컬럼명 리스트로 정의
drop_cols = ['매수자', '매도자', '해제사유발생일', '거래유형', '중개사소재지', '등기일자']
```

(4) 불필요한 컬럼 제거 후 csv 저장

```python
# 정의한 컬럼 제거 후 새 csv 저장

df = df.drop(columns=drop_cols)
df.to_csv(r"C:\Users\user22\Desktop\아파트 값 예측\new_city2.csv", index=False, encoding='utf-8-sig')

print(f"컬럼 제거 후: {df.columns.tolist()}")
```
